```python
import pandas as pd
import numpy as np
# import plotly
# import plotly.figure_factory as ff
# import plotly.express as px
# import plotly.graph_objects as go
import json

```

```python
election_df = pd.read_csv("1976-2016-president.csv")
```

```python
cleanestelection_df = election_df.query('party == "democrat" | party == "republican"')
cleanestelection_df
```

```python
clean_electiondf = cleanestelection_df[['year', 'state', 'state_po', 'state_fips', 'state_cen', 'state_ic',
        'candidate', 'party', 'candidatevotes',
       'totalvotes']]
clean_electiondf.head(45)
```

```python
dem_electiondf = election_df.query('party == "democrat"')
rep_electiondf = election_df.query('party == "republican"')
combined_electiondf = pd.merge(dem_electiondf, rep_electiondf,  on =['state', 'year']) 

combined_electiondf
```

```python
combined_electiondf['winner1'] = np.where(combined_electiondf['candidatevotes_x'] > combined_electiondf['candidatevotes_y'], 'blue', 'red')
```

```python
combined_electiondf
```

```python
df_1976 = combined_electiondf[(combined_electiondf['year']== 1976 )]
df_1976
```

```python
geojson_source = "states.json"
geojson_pd = pd.read_json(geojson_source, encoding='latin-1', lines=True)
geojson_pd.head()

doublecombined_electiondf = combined_electiondf.merge(geojson_pd, left_on='state_fips_x', right_on='id')

# doublecombined_electiondf[''.astype(int)
doublecombined_electiondf
```

```python
def df_to_geojson(df, properties):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{}
#                    'geometry':{}
                  }
        feature['geometry'] = row['geometry'] 
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
properties = ['year' ,'state_po_x',  'candidate_x', 'candidatevotes_x', 'totalvotes_x', 'candidate_y' ,'candidatevotes_y', 'winner1']
geojson = df_to_geojson(doublecombined_electiondf, properties)
```

```python
geojson
with open('mydata.json', 'w') as f:
    json.dump(geojson, f)
```
